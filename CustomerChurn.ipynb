Till now we have loaded the data. now we will start with importing all the necessary libraries and then we will explore and preprocess the data.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report
```

Now we will split the data into training and testing sets assuming that our target variable is 'churn'.

```python

X = data.drop('Churn', axis=1)
y = data['Churn']

non_numeric_cols = X.select_dtypes(exclude=['number']).columns


X_encoded = pd.get_dummies(X, columns=non_numeric_cols)


X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)


model = LogisticRegression()
model.fit(X_train, y_train)
```

Now we will evaluate the model's performance.

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
confusion_mat = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(confusion_mat)
```

Now we will define a function to predict customer churn.

```python
import pandas as pd

def predict_customer_churn(customer_data, model):


    feature_names = model.feature_names_in_


    input_data = pd.DataFrame(0, index=[0], columns=feature_names)


    for feature, value in customer_data.items():
        if feature in feature_names:
            input_data[feature] = value

    churn_prediction = model.predict(input_data)[0]

    if churn_prediction == 1:
        return "Customer will churn"
    else:
        return "Customer will not churn"


print("Please enter the customer's information:")

tenure = int(input("Tenure (in months): "))
monthly_charge = float(input("Monthly charge: "))
contract_type = int(input("Contract type (0 for month-to-month, 1 for annual): "))
gender = input("Gender (Male/Female): ")
age = int(input("Age: "))

customer_data = {
    'tenure': tenure,
    'monthly_charge': monthly_charge,
    'contract_type': contract_type,
    'gender': gender,
    'age': age
}

churn_prediction = predict_customer_churn(customer_data, model)
print(churn_prediction)
```
